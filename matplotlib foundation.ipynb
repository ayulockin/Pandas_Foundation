### Imports

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

### Simple Plotting

Look for title and x and y axis labels.

```python
x = [1,2,3]
y = [5,2,6]
```

```python
plt.plot(x,y);
plt.xlabel('x values');
plt.ylabel('y values');
plt.title('graph');
plt.tight_layout()
```

### Legends

```python
y2 = [10, 13, 11]
```

```python
plt.plot(x, y, label='line1')
plt.plot(x, y2, label='line2')
plt.legend(loc='upper left', frameon=False)
```

### Sine and Cosine

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure()
plt.plot(x, y1, '-', label='sin');
plt.plot(x, y2, '--', label='cos');
plt.legend();
```

### Save Plots

```python
fig.savefig('my_figure.png')
```

### Subplots

##### The MATLAB way

```python
plt.figure()

plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, y1)

plt.subplot(2, 1, 2)
plt.plot(x, y2);
```

##### The Pythonic way

```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

### Control colors

```python
plt.plot(x, np.sin(x - 0), color='blue')    
plt.plot(x, np.sin(x - 1), color='g')           
plt.plot(x, np.sin(x - 2), color='0.75')        
plt.plot(x, np.sin(x - 3), color='#FFDD44')     
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) 
plt.plot(x, np.sin(x - 5), color='chartreuse'); 
```

### Control line type

```python
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

### Control X-axis and Y-axis range

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

### Scatter Plot

```python
plt.scatter(x, np.sin(x), marker='o');
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

### Histograms

### More Plots

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 2, figsize=(12,3))

axes[0].step(n, n**2, lw=2)
axes[0].set_title("step")

axes[1].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[1].set_title("bar")

plt.tight_layout()
```
